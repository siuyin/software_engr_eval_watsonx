```python
from ibm_watsonx_ai.foundation_models import Model
import os,getpass

def get_credentials():
    return {
        "url" : "https://us-south.ml.cloud.ibm.com",
        "apikey" : os.getenv("API_KEY")
    }

def summarize(review_df):
    review_text=review_df.to_csv(index=False)
    swe_name = review_df['Software Engineer'].values[0]
    swe_features=list(map(str,review_df['Feature ID'].values))
    swe_feature_ids = ",".join(swe_features)

    if swe_name.lower() == 'alex lee' or swe_name.lower() == 'emily chen':
        return
    
    parameters = {
        "decoding_method": "greedy",
        "max_new_tokens": 200,
        "min_new_tokens": 25,
        "stop_sequences": ["\n\n"],
        "repetition_penalty": 1
    }
    
    model = Model(
        model_id = "ibm/granite-13b-chat-v2",
        params = parameters,
        credentials = get_credentials(),
        project_id = os.getenv('PROJECT_ID')
        )
    
    prompt_input = f"""You are an Engineering Manager preparing to evaluate the peformance of your sofware engineers. Your aim is to write a one paragraph summary of their work.
    
    Feature ID,Software Component,Feature,Software Engineer,Engineers' Comments
    1,Authentication Module,Enhanced Security,Alex Lee,Improved password hashing and salt generation.
    3,Database Optimization,Performance Boost,Alex Lee,Optimized database queries and indexes.
    5,Error Handling,Robust Recovery,Alex Lee,Implemented comprehensive error handling and logging.
    Alex Lee's work focused on enhancing security, performance, and error handling. He made significant improvements to the authentication module by improving password hashing and salt generation, optimized database queries and indexes for database optimization, and implemented comprehensive error handling and logging for error handling.



    Feature ID,Software Component,Feature,Software Engineer,Engineers' Comments
    2,User Interface,Modernized Design,Emily Chen,Redesigned UI for better user experience.
    4,API Integration,Seamless Connectivity,Emily Chen,Integrated with third-party APIs for extended functionality.
    6,Search Functionality,Advanced Filtering,Emily Chen,Added advanced search filters for efficient data retrieval.
    Emily Chen's work concentrated on modernizing the user interface, integrating with third-party APIs, and adding advanced search filters. She redesigned the UI for a better user experience, integrated with third-party APIs for extended functionality, and added advanced search filters for efficient data retrieval.

    
    {review_text}
    """
    
    print("\n\n-----------------")
    generated_response = model.generate_text(prompt=prompt_input, guardrails=True)
    print(f"{swe_name} implemented {len(swe_features)} features: [{swe_feature_ids}] in the review period.")
    print(generated_response)

```

```python
for swe in sw_engrs:
    sw_df=sw_engr_df[sw_engr_df['Software Engineer']==swe]
    summarize(sw_df)
```
